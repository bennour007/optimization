```python
import pandas as pd
import numpy as np
import mlrose
```

```python
df = pd.read_csv("/home/bennour/Downloads/tunisia_cities - data.csv")
cities = df.columns.drop('City')
```

```python
df.fillna(value=0, inplace=True)
del df['City']
```

```python
df.columns = np.arange(len(df.columns))
```

```python
df.insert(0, column = 'city', value = np.arange(14))
```

In this report we will study the traveling salesman probelm, from the perspective of a politician who wants to visit the cities proposed in the table below, and he wants to minimize the total distance. 
We consider this problem as symetric for the sake of simplicity of impliementation. 

```python
df_1 = pd.melt(df, id_vars= df.columns[0], var_name='distination')
df_1 = df_1[df_1.value > 0]
df_1
```

```python
# creating a list of couples 
df_2 = df_1.values.tolist()
```

```python
print(df_2)
```

```python
fitness_dists = mlrose.TravellingSales(distances=df_2)
```

```python
problem_fit = mlrose.TSPOpt(length= 14,
                           fitness_fn= fitness_dists, 
                           maximize=False)
```

```python
schedule = mlrose.ExpDecay()
```

```python
best_state, best_fitness = mlrose.simulated_annealing(problem_fit,
                                                     random_state=1,
                                                      max_attempts=182,
                                                      max_iters= 10000,
                                                      schedule=schedule)
```

```python
print('The best state found is: ', best_state, '\n', cities[best_state])

print('The fitness at the best state is: ', best_fitness)
```
